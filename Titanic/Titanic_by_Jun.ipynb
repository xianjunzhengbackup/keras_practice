```python
import seaborn as sns
import matplotlib.pyplot as plt
```

下面是个好方法显示散点图。seaborn是MIT的库 <br>
sns.FacetGrid决定生成几幅图，Emabarked 有3种选项，sex有两种选项，所以将会有6副图。图中散点的颜色有hue决定，seismic只是调色板 <br>
grid.map决定了散点的横轴与纵轴

```python
grid = sns.FacetGrid(pd_train, col = "Embarked", row = "Sex", hue = "Survived", palette = 'seismic')
grid = grid.map(plt.scatter, "PassengerId", "Age")
grid.add_legend()
grid
```

```python
grid = sns.FacetGrid(pd_train, hue = "Survived",palette='seismic')
grid=grid.map(plt.scatter, "PassengerId",'Age')
grid.add_legend()
grid
```

```python
grid = sns.FacetGrid(pd_train, col = "Pclass", row = "Sex", hue = "Survived", palette = 'seismic')
grid = grid.map(plt.scatter, "PassengerId", "Age")
grid.add_legend()
grid
```
